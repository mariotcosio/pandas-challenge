```python
# Import Modules
import pandas as pd
import numpy as np

# Establish file path for csv file
csv_file = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(csv_file)

purchase_data.head()
```

```python
# Display the total number of players
total_players = purchase_data["SN"].drop_duplicates()
total_players = total_players.count()
total_players
pd.DataFrame({"Total Players": [total_players]})
```

```python
#  Purchasing Analysis (Total)
 
# Run basic calculations to obtain number of unique items, average price, etc.
# Create a summary data frame to hold the results, Optional: give the displayed data cleaner formatting,
# Display the summary data frame
```

```python
unique_items = purchase_data["Item Name"].nunique()
avg_price = round(purchase_data["Price"].mean(), 2)
purchase_id = purchase_data["Purchase ID"].count()
total_rev = purchase_data["Price"].sum()

df = pd.DataFrame([{"Unique Items": unique_items, "Average Purchase Price": avg_price, "Total Purchases":
                   purchase_id, "Total Revenue": total_rev}])
df
```

```python
# Gender Demographics
# Percentage and Count of Male Players, Percentage and Count of Female Players, Percentage and Count of Other / Non-Disclosed
```

```python
gender = purchase_data.groupby(["Gender"])
percent_gender = round((gender["SN"].count() / total_players) * 100, 2)

gender_df = pd.DataFrame({"Player Count": gender["SN"].count(), "Percent of Players": percent_gender})
gender_df
```

```python
# Purchasing Analysis (Gender)
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
```

```python
gender_groupby = purchase_data.groupby(["Gender"])
purchase_count_gender = gender_groupby["Price"].count()
avg_price_gender = round(gender_groupby["Price"].mean(), 2)
total_value_gender = round(gender_groupby["Price"].sum(), 2)
per_person =  total_value_gender / gender_df["Player Count"]
purchase_df = pd.DataFrame({"Purchase Count": purchase_count_gender, "Avg. Purchase Price": avg_price_gender,
                           "Total Purchase Value": total_value_gender, "Total Purchase per Person": per_person})
purchase_df["Avg. Purchase Price"] = purchase_df["Avg. Purchase Price"].map("${:.2f}".format)
purchase_df["Total Purchase Value"] = purchase_df["Total Purchase Value"].map("${:.2f}".format)
purchase_df["Total Purchase per Person"] = purchase_df["Total Purchase per Person"].map("${:.2f}".format)
purchase_df
```

```python
# Age Demographics
# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table
```

```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-40", "40+"]
player_demo = purchase_data.loc[:, ["Age", "Gender", "Price"]]
player_demo["Age Ranges"] = pd.cut(player_demo["Age"], bins, labels=labels)
```

```python
age_totals = player_demo["Age Ranges"].value_counts()
age_percentage = round((age_totals / total_players *100), 2)
age_demo = pd.DataFrame({"Total Players": age_totals, "Percent of Players": age_percentage})
age_demo["Percent of Players"]= age_demo["Percent of Players"].map("{:.2f}%".format)
age_demo = age_demo.sort_index()
age_demo
```

```python
# Purchasing Analysis (Age)
# Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
```

```python
bins = [0, 10, 15, 20, 25, 30, 35, 100]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-40"]
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=labels)
purchase_data_groupby = purchase_data.groupby(by = "Age Ranges")
purchase_count = purchase_data_groupby["Price"].count()
avg_purchase_price = round(purchase_data_groupby["Price"].mean(), 2)
total_purchase = round(purchase_data_groupby["Price"].sum(), 2)
avg_per_person = round(total_purchase / purchase_count, 2)
purchase_analysis = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price":
                                 avg_purchase_price, "Total Purchase Price": total_purchase, 
                                 "Average Per Person": avg_per_person})
purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis["Total Purchase Price"] = purchase_analysis["Total Purchase Price"].map("${:.2f}".format)
purchase_analysis["Average Per Person"] = purchase_analysis["Average Per Person"].map("${:.2f}".format)
purchase_analysis = purchase_analysis.sort_index()
purchase_analysis
```

```python
# Top Spenders
# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame
```

```python
total_by_user = purchase_data.groupby(["SN"]).sum()["Price"]
avg_by_user = round(purchase_data.groupby(["SN"]).count()["Price"], 2)
count_by_user = purchase_data.groupby(["SN"]).count()["Price"]
top_spenders = pd.DataFrame({"Total Purchase Value": total_by_user, "Average Purchase Price": avg_by_user,
                            "Purchase Count": count_by_user})
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders.sort_values("Total Purchase Value", ascending = False).head()
```

```python
# Most Popular Items
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
# Create a summary data frame to hold the results
# Sort the purchase count column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame
```

```python
popular_items_group = purchase_data.groupby(["Item ID", "Item Name"])
popular_purchase_count = popular_items_group['Price'].count()
popular_price = popular_items_group['Price'].mean()
total_price = popular_items_group["Price"].sum()

df_popular_items = pd.DataFrame({'Purchase Count': popular_purchase_count, "Total Purchase Value": total_price, "Item Price": popular_price})
df_popular_items["Item Price"] = df_popular_items["Item Price"].map("${:.2f}".format)
df_popular_items.sort_values("Purchase Count", ascending=False).head()
```
